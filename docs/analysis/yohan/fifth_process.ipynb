```python
# Okt 형태소 분석기 인스턴스 생성
okt = Okt()

import re
import numpy as np


# 토크나이징 함수 정의
def tokenizer(raw, pos=["Noun","Alpha","Verb","Number"], stopword=stopword):

  get_list = []

  for word, tag in okt.pos(raw,norm=True,stem=True):
    if len(word) <= 1:
      word = '공란'

    if tag in pos and word not in stopword:
      get_list.append(word)
    else:
      get_list.append('공란')
    pass


  return get_list
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=tokenizer, use_idf=True)
```

- LDA 토픽 추출 함수 제작

```python
def LDAs(vectorizer,input_text_list):


  features = vectorizer.fit_transform(input_text_list)

  # 단어 사전 추출
  dictionary_list = vectorizer.get_feature_names_out()

  # LDA 학습
  from sklearn.decomposition import LatentDirichletAllocation
  lda_model = LatentDirichletAllocation(n_components=5, random_state=55) #인스턴스화 #n_components 토픽의 갯수
  lda_model.fit(features) #교육

  ## 상위 단어 추출
  ## 0 확률 1은 dictionary
  topics_list = list()
  for topic in lda_model.components_:
      df_datas = [topic, dictionary_list]
      df_topics = pd.DataFrame(data=df_datas)
      df_topics= df_topics.T
      df_topics = df_topics.sort_values(0, ascending=False)

      topics_text = ' '.join(df_topics[1].values[:10])# 시리즈 형식으로 출력 get values from series / index
      topics_list.append(topics_text)

  return topics_list
```

```python
get_topics = []
counter = 0
for i in dfs:
  # db저장시 발생했던 고유값 제거
  i = i.drop(columns = '_id')

  # 라벨링 함수 적용
  i['textLevel'] = i['textLevel'].apply(labeling)

  # 긍정문 추출
  processing_df = i.query('textLevel == 1')

  # 토픽 추출
  try:
    get_topics.append(LDAs(vectorizer,processing_df['textContent']))
  except:
    get_topics.append(f'오류 발생 : {counter}번재 csv')

  counter += 1
```

- 군집화 최적 코드를 작성 -> 실루엣 계수 활용

```python
from sklearn.metrics import silhouette_score
```

```python
def silhouette_point(vectorizer,input_text_list):
  k_range = range(2,11)

  best_n = -1
  best_score = -1
  best_score_list = []
  for k in k_range:

      features = vectorizer.fit_transform(input_text_list)

      from sklearn.decomposition import LatentDirichletAllocation

      lda_model = LatentDirichletAllocation(n_components=k, random_state=55) #인스턴스화 #n_components 토픽의 갯수
      lda_model.fit(features) #교육

      clusters = lda_model.transform(features).argmax(axis=1)  # 각 문서의 클러스터 할당

      score = silhouette_score(features,clusters)
      print('k :',k,'score :',score)

      if score > best_score:
          best_n = k
          best_score = score

  print('best n :', best_n, 'best score :', best_score)
  best_score_list.append(best_n)

  return best_score_list
```

```python
best_score = []
counter = 0
for i in dfs:
  # db저장시 발생했던 고유값 제거
  i = i.drop(columns = '_id')

  # 라벨링 함수 적용
  i['textLevel'] = i['textLevel'].apply(labeling)

  # 긍정문 추출
  processing_df = i.query('textLevel == 1')

  # 토픽 추출
  try:
    best_score.append(silhouette_point(vectorizer,processing_df['textContent']))
  except:
    best_score.append(f'오류 발생 : {counter}번재 csv')

  counter += 1
```
